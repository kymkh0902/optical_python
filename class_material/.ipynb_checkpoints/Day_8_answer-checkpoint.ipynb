```python
# 예제1) import 및 그래프 출력 설정

#1. matplotlib package에서 pyplot을 plt라는 이름으로 import합니다. 
import matplotlib.pyplot as plt

#2. jupyter에서 그래프를 출력하도록 설정하세요  
%matplotlib inline
```

```python
# 예제2) 그래프를 그릴 pandas 데이터 가져오기

names = ['날짜','Lot','Frame수', '카메라 번호', '사용안함', 'INDEX', '불량유형','X','y', 
         'X-Size','Y-Size', '불량이미지','사용안함','사용안함','Value','흑점','조명밝기']

#1. os 를 import 한 후에 dataset를 작업 폴더로 지정합니다. 20160715DC02017.xlsx 파일을 불러옵니다. (변수 : data)
import os
os.chdir('../dataset')
import pandas as pd

data = pd.read_excel('20160715DC02017.xlsx', names = names)

#2. na값을 drop하고 '강/약' column을 만듭니다. (불량 유형을 60000 기준으로 작으면 강, 크면 약, lambda 함수 쓰세요)
data.dropna(inplace = True)
data['강/약'] = data['불량유형'].apply(lambda x: '강' if x<60000 else '약')

#3. Size라는 열을 만들어줍니다. (X-Size 와 Y-Size의 평균 값입니다.)
data['Size'] = (data['X-Size'] + data['Y-Size'])/2

#4. data에서 크로스2 휘점에 해당되는 값들만 불러와서 data_cross2라는 변수에 넣습니다. (불량 유형 : 257, 65793, 사용 함수 : isin)
data_cross2 = data[data['불량유형'].isin([257, 65793])]
```

```python
# 예제3) 산점도 그래프 그리기. (사용할 데이터는 계속 data_cross2입니다.)

#1. subplots 함수로 원하는 크기의 그래프를 하나 만들어보세요. (변수 : fig, ax) 그리고 출력해보세요 뭐가 나오나요?
fig, ax = plt.subplots(figsize = (8,6))

#2. scatter plot을 그려보겠습니다. (x : Value, y : Size) 크기, 색깔 등을 설정해보세요
ax.scatter(x = data_cross2['Value'], y = data_cross2['Size'], s = 15, linewidths = 1, # x: x축 값, y: y축 값, s: 점 크기, linewidth: 점 테두리 두께 
           facecolors = 'none', edgecolors = 'r', alpha = 0.5) # facecolors: 점 색, edgecolors: 테두리 색, alpha: 투명도

#3. x축, y축 설정을 변경해보겠습니다. 원하시는 그래프를 생각하시고 하나씩 설정을 변경해보세요. 
ax.set_xlim(0,data_cross2['Value'].max()) # x축 출력 범위 입력
ax.set_ylim(0,data_cross2['Size'].max()) # y축 출력 범위 입력
ax.set_xticks([i for i in range(0, 200, 10)]+[int(data_cross2['Value'].max())])  # x축 label 설정
ax.set_yticks([i for i in range(0, int(data_cross2['Size'].max()), 1)]+[int(data_cross2['Size'].max())]) # y축 label 설정
ax.legend(['Marking(Cross2)']) # 범례 추가
ax.tick_params(axis = 'both', which = 'both', length = 0) # 축 표시해주는 tick 설정. axis : 축, which : tick 종류, length : tick 길이 
ax.spines['top'].set_visible(False) # 그래프 둘러싼 선에 대한 설정. top, bottom, left, right로 가능. 
ax.grid('on') # grid on/off 

plt.show()

#4. 본인이 더 고쳤으면 좋겠다고 생각하는 설정을 말씀해주세요. (모든 설정이 변경 가능합니다.)
```
